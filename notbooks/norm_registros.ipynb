```python
import pandas as pd
```

```python
df = pd.read_csv('../data/indicador_bienestar.csv', encoding='latin-1')
df
```

```python
df.rename(columns={'valores y media': 'nv_general',
                   'características sociodemográficas': 'caract',
                   'Satisfacción con la vida': 'satis_vida',
                   'Valor de la vida': 'valor_vida',
                   'Satisfacción con el tiempo de ocio': 'satis_tmp_ocio',
                   'Satisfacción con la vivienda y el entorno': 'satis_vivienda',
                   'Satisfacción con la economía doméstica': 'satis_economica',
                   'Estado de ánimo': 'animo',
                   'Relaciones personales': 'relaciones',
                   'Confianza en las personas': 'confi_personas',
                   'Confianza en los poderes públicos': 'confi_politica'
                    },
          inplace=True)
```

```python
df.drop(columns='periodo', inplace=True)
```

```python
df = df[df['nv_general'] != 'MEDIA']
```

```python
print(df['nv_general'].unique())
mapeo = {'Bajo': 0, 'Medio': 1, 'Alto': 2}
```

```python
df.loc[:, 'nv_general'] = df['nv_general'].map(mapeo)
df
```

```python
df['sexo'] = df['sexo'].map({'Hombre': 0, 'Mujer': 1})
df
```

```python
df.loc[:, 'caract'] = df['caract'].str.replace(r'^-', '', regex=True)
df
```

```python
df.loc[:, 'caract'] = df['caract'].str.replace(r'/.*|\(.*', '', regex=True).str.strip()
df
```
